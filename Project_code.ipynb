Data is divided on the basis of Hours. To reduce the size of data

```python
Df_resample_hourly = df.resample('H').mean()
```

The data is scaled to save some computation time

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled = scaler.fit_transform(Df_resample_hourly)
```

# Train Test Split


The data is divided such that the training set comprises of initial 2 years and test set contains the remaining 1 year. Such that we Test our model for future predictions, rather than randomly picking Test Set in between Training Set

```python
n_train_time = 365*24*3
train = scaled[:n_train_time, :]
test = scaled[n_train_time:, :]
```

```python
train.shape
```

```python
test.shape
```

Train - Test - Split

```python
X_train = train[:,1:7]
y_train = train[:,0]
X_test, y_test = test[:, 1:7], test[:,0]
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
train.shape
```

```python
y_train
```

# Model Creation

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model = Sequential()

model.add(Dense(6, activation = 'relu'))
model.add(Dense(6, activation = 'relu'))
model.add(Dense(6, activation = 'relu'))
model.add(Dense(6, activation = 'relu'))

model.add(Dense(1))

model.compile (optimizer = 'adam',loss = 'mse')
```

# Fiting the model and implying Early Stopping

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', mode = 'min', verbose = 1, patience = 25)
```

```python
model.fit(x = X_train, y = y_train, epochs = 200,validation_data= (X_test,y_test),
         batch_size= 100,callbacks = [early_stop])
```

# Plotting The losses

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

# Performance

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_squared_error(y_test, predictions)**0.5
```

```python
mean_absolute_error(y_test,predictions)
```

```python
explained_variance_score(y_test,predictions)
```
